```python
def agrton_to_float(x,loc):
    x = x.split("/")[loc].replace(" ","")
    try:
        x= int(x)
    except:
        x= np.nan
    return(x)
```

```python
# df_analise['agtron_lb'] = df_analise['agtron'].apply(lambda row: row.split("/")[0].replace(" ","")) #limite inferior
# df_analise['agtron_ub'] = df_analise['agtron'].apply(lambda row: row.split("/")[-1].replace(" ","")) #limite superior

df_analise['agtron_lb'] = df_analise['agtron'].apply(lambda row: agrton_to_float(row,0)) #limite inferior
df_analise['agtron_ub'] = df_analise['agtron'].apply(lambda row: agrton_to_float(row,-1)) #limite superior
```
